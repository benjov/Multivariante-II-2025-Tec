# Calculo de deciles de ingreso

```python
# -------------------------
# 1) Cuantiles ponderados
# -------------------------
def weighted_quantiles(values, weights, qs):
    """
    values: array de valores (ingresos)
    weights: array de ponderadores (>0)
    qs: iterable de cuantiles en [0,1], p.ej. [0.1, 0.2, ..., 0.9]
    Retorna los cortes (valores de 'values') donde la CDF ponderada alcanza cada q.
    """
    values = np.asarray(values, dtype=float)
    weights = np.asarray(weights, dtype=float)
    assert np.all(weights >= 0), "Hay ponderadores negativos."

    # ordenar por valor
    sorter = np.argsort(values)
    v = values[sorter]
    w = weights[sorter]

    # CDF ponderada
    cw = np.cumsum(w)
    if cw[-1] == 0:
        raise ValueError("La suma de ponderadores es 0.")
    cdf = cw / cw[-1]

    # para cada q, primer ndice con cdf >= q
    qs = np.asarray(qs, dtype=float)
    idx = np.searchsorted(cdf, qs, side="left")
    idx = np.clip(idx, 0, len(v)-1)

    return v[idx]
```

```python
# -----------------------------------------------------
# 2) Deciles ponderados (cortes) y asignacin a filas
# -----------------------------------------------------
def add_income_deciles(
    df, value_col="ing_cor", weight_col="factor",
    out_col="decil_ing"
):
    """
    Calcula cortes declicos ponderados para toda la muestra
    y asigna a cada rengln su decil (1..10).
    Devuelve (df_con_decil, cortes_series).
    """
    # Filtrar filas vlidas para calcular cortes
    mask = df[value_col].notna() & df[weight_col].notna() & (df[weight_col] > 0)
    if mask.sum() == 0:
        raise ValueError("No hay observaciones vlidas con ingreso y factor > 0.")

    qs = np.linspace(0.1, 0.9, 9)  # 10%, 20%, ..., 90%
    cuts = weighted_quantiles(df.loc[mask, value_col], df.loc[mask, weight_col], qs)

    # Construir bins e imputar decil
    bins = np.concatenate(([-np.inf], cuts, [np.inf]))
    labels = list(range(1, 11))  # 1..10

    deciles = pd.cut(df[value_col], bins=bins, labels=labels, include_lowest=True)
    df[out_col] = deciles.astype("Int64")  # permite NA si ing_cor es NA

    cortes = pd.Series(cuts, index=[f"q{int(q*100)}" for q in qs], name="corte")
    return df, cortes
```

```python
#
df, cortes = add_income_deciles(df, value_col="ing_cor", weight_col="factor", out_col="decil_ing")
print(cortes/3)
```

```python
#3) Chequeo rpido: participacin ponderada por decil (debera ~10% cada uno)
share = df.dropna(subset=["decil_ing"]).groupby("decil_ing")["factor"].sum() / df["factor"].sum()
print(share)
```

```python
# Calculate the average income for each income decile
# This assumes that the 'df' DataFrame has a 'decil_ing' column with the assigned deciles
average_income_by_decile = df.groupby('decil_ing')['ing_cor'].mean()

average_income_by_decile/3
```
